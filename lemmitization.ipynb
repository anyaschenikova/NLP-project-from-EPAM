```python
import nltk
import pandas as pd

tokenizer = nltk.tokenize.WordPunctTokenizer()
```

```python
data = pd.read_csv("/content/dataset.csv")
data.head()
```

```python
data.columns
```

```python
data2 = pd.concat([data['Header'], data['Text']], ignore_index=True)
data2 = pd.concat([data2, pd.Series(data['Category'].unique())], ignore_index=True)
data_new = pd.DataFrame(data2).dropna()
print(data_new)
```

```python
words = []
for string in data_new[0]:
  tokens = tokenizer.tokenize(string)
  [words.append(word) for word in tokens]
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('russian')
```

```python
filtered_tokens = []
for token in tokens:
  if token not in stop_words:
    filtered_tokens.append(token)
```

```python
import pymorphy2

def lemmitization(words):
  morph = pymorphy2.MorphAnalyzer()
  words_lem = []
  print('Still working...')
  for word in words:
    words_lem.append(morph.parse(word)[0].normal_form)
  df = pd.DataFrame(words_lem)
  df.to_csv('lemma.csv', encoding = 'utf8')
  print('success')
```
